BUILDING THE MACHINE LEARNING MODEL USING LINEAR REGRESSION

```python
#import machine learning libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
```

```python
#display the features and target variables for splitting
X = df_fil.drop(['SalePrice'], axis = 1)
y = df_fil['SalePrice']
```

```python
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2) #splitted the dataset
```

```python
scale = StandardScaler()
X_train_scaled = scale.fit_transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train) #fit the linear regression model
```

```python
y_pred = model.predict(X_test_scaled) #predict the target variable using the test dataset
y_pred
```

```python
log_y_test = np.log(y_test) #calculated the logarithmic value of predicted value and the observed sales price.
log_y_pred = np.log(y_pred)
```

```python
mae = mean_absolute_error(log_y_test, log_y_pred) #calculated the metric - Mean Absolute Error
mae
```

```python
mse = mean_squared_error(log_y_test, log_y_pred) #calculated the metric - Mean Squared Error
rmse = np.sqrt(mse)
rmse
```

```python
alpha = 1.0  # Regularization strength (hyperparameter)
lasso = Lasso(alpha=alpha)
lasso.fit(X_train_scaled, y_train)

# Make predictions
y_pred_test = lasso.predict(X_test_scaled)
y_pred_test
```
