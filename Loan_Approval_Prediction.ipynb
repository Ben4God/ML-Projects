# Stacking Method

```python
!pip install scikit-learn xgboost
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import StackingClassifier
```

```python
# Define base models
level0 = [
    ('lr', LogisticRegression()),  # Logistic Regression
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),  # Random Forest
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))  # XGBoost
]
```

```python
# Define meta-model (Logistic Regression in this case)
level1 = LogisticRegression()
```

```python
# Stacking classifier with base models and a meta-model
model5 = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
```

```python
model5.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model5.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Stacking Model Accuracy: {accuracy:.4f}')
```
